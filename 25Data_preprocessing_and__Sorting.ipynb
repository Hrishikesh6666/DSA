```python
df = pd.read_csv('/content/Final.csv',index_col=False)
```

# SAVE THE DATA INTO INTIGRATED FORMATE

```python
lst=[]
for i in df['eng']:
 lst.append(int(i))
 print(lst)
```

```python
def typecast(col_name):

    lst = []
    for i in df[col_name]:
        lst.append(int(i))
    df[col_name] = lst
```

```python
typecast('eng')
typecast('hin')
typecast('mth')
typecast('sci')
```

```python
df
```

#ADDING NEW DATA INTO THE DATASET (AVG/TOTAL)

```python
df.iloc[1,1:].sum()
```

```python
total =[]
avg = []
for i in range (len(df)):
  total.append(df.iloc[i,1:].sum())
  avg.append(df.iloc[i,1:].sum()/5)
```

```python
df['total']=total
df['avg']= avg
df
```

# SORTING STUDENT DATA BASED ON MARKS

```python
df_pr = df.sort_values(by='total',ascending=False) #topper of the class
```
